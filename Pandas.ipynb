**In-Depth**



1.   Pandas is an open source data analysis library written in python
2.   It leverage the power and speed of numpy to make data analysis and preprocessing easy for data scientists

3.   It provides rich and highly robust data operations








*   Pandas has two type of data structures:

  1.   **Series** - It's one dimensional array with indexes, it stores a single column or row of data in a Dataframe
  2.   **Dataframe** - It's a tabular spreadsheet like structure representing rows each of which contains one or multiple columns


*   A one-dimensional array(labeled) capable of holding any type of data-series

*   A two-dimensional data(labeled) structure with columns of potentially different type of data - DataFrame

```python
df = pd.DataFrame({
    'name':['Alfred','Batman','Alfred'],
    'toy':[np.nan,np.nan,np.nan],
    'born':[pd.NaT,pd.Timestamp('1940-04-25'),pd.NaT]
})
```

```python
df
```

```python
df.dropna(how='all',axis=1) #Remove missing values.
```

```python
df.head() #Return the first n rows.
```

```python
df.drop_duplicates(subset=['name'],keep='first')
```

```python
df.drop_duplicates(subset=['name'],keep='last')
```

```python
df.drop_duplicates(subset=['name'],keep=False) #Return DataFrame with duplicate rows removed.
```

```python
df.shape #Return a tuple representing the dimensionality of the DataFrame.
```

```python
df.info #Print a concise summary of a DataFrame.
```

```python
df.info() 
```

```python
df['toy'].value_counts(dropna=False)
```

```python
df['name'].value_counts(dropna=False) #dropna - Dont include counts of NaN.
```

```python
df['toy'].value_counts(dropna=True) #Return a Series containing counts of unique values.
```

```python
df.notnull() #Detect non-missing values for an array-like object.
```

```python
df.isnull() #Detect missing values for an array-like object.
```
